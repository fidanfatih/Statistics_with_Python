```python
import random
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
random.seed(101)
```

#### Example-1

```python
# H0: mu=10 # null hypothesis, (we can say H0: mu ≤ 10 )
# H1: mu>10 # alternative hypothesis, right tail
mu0=10
sigma=1.5
n=40
x=10.5
alpha=0.05 #significant level
```

```python
# n >30, sigma is known, observations are independent so we use z test.
z= (x-mu0)/(sigma/sqrt(n))
z
```

```python
pV=1-stats.norm.cdf(z) # p-Value
pV
```

```python
# P( z>2.1 | H0=True ) =0.0179

# reject the null hypothesis(H0) when p-Value ≤ alpha
# pV < alpha, so We reject the null hypothesis
```

```python
# Critical Value=?
# z= (x-mu)/(sigma/sqrt(n)) >> Xc = mu0 + Zα*(sigma/sqrt(n))

mu0 + stats.norm.ppf(1-alpha)*(sigma/sqrt(n))
```

```python
# we should accept alternative hypothesis. So we can say the beach is not safe.
```

```python
# H0: x2=x1 # null hypothesis
# H1: x2<x1 # alternative hypothesis, (left tail)
n1=10
n2=10
x1=121
x2=112
s1=s2=8 # Standard Error of Sample
alpha=0.05 #significant level
delta=0

# We use t-test here. 
```

```python
T0 =(x2-x1-delta)/(Sp*sqrt(1/n1 + 1/n2)) # Test statistics
T0
```

```python
df =n1+n2-2
pV= stats.t.cdf(T0,df) # p value
pV
```

```python
if pV < alpha:
    print('At {} level of significance, we can reject the null hypothesis in favor of Ha.'.format(alpha))
else:
    print('At {} level of significance, we fail to reject the null hypothesis.'.format(alpha))
```

#### Example-9

```python
data = pd.read_csv("catalysts.csv")
data
```

```python
# Difference in means of two normal distributions, variances are unknown and equal...

# H0: x1-x2 = 0 # null hypothesis
# H1: x1-x2 > 0 # alternative hypothesis
n1=len(data['Catalyst1'])
n2=len(data['Catalyst2'])
x1=data['Catalyst1'].mean()
x2=data['Catalyst2'].mean()
s1=data['Catalyst1'].std()
s2=data['Catalyst2'].std()
alpha=0.05 #significant level
delta=0
```

```python
print('x1 = {:.3f}'.format(x1))
print('x2 = {:.3f}'.format(x2))

print('s1 = {:.3f}'.format(s1))
print('s2 = {:.3f}'.format(s2))
```

```python
Sp =sqrt( ( (n1-1)*s1**2+(n2-1)*s2**2 ) /(n1+n2-2) ) # S pooled
print('s_pooled = {:.3f}'.format(Sp))
```

```python
T0 =(x1-x2-delta)/(Sp*sqrt(1/n1 + 1/n2)) # Test statistics
print ('t_statistic = {:.3f}'.format(T0))
```

```python
df =n1+n2-2
pV= 2*stats.t.cdf(T0,df) # p value, two tailed
print ('p_value = {:.3f}'.format(pV))
```

```python
# reject the null hypothesis(H0) when p-Value ≤ alpha
# pV > alpha, so fail to reject the null hypothesis
```

```python
if pV < alpha:
    print('At {} level of significance, we can reject the null hypothesis in favor of Ha.'.format(alpha))
else:
    print('At {} level of significance, we fail to reject the null hypothesis.'.format(alpha))
```

```python
# we dont have strong evidence to conclude that catalist 2 results in a mean yield 
# that differs from the mean yield when catalist 1 is used.
```

```python
# OR SOLUTION WITH Scipy.stats.ttest
```

```python
(T0, pV) = stats.ttest_ind(data['Catalyst1'], data['Catalyst2'], equal_var = True)
print ('t_statistic = {:.3f}'.format(T0))
print ('p_value = {:.3f}'.format(pV))
```

```python
if pV < alpha:
    print('At {} level of significance, we can reject the null hypothesis in favor of Ha.'.format(alpha))
else:
    print('At {} level of significance, we fail to reject the null hypothesis.'.format(alpha))
```
