#### EXERCISE 1. 
Create a Standard Normal Distribution Table using Python scipy.stats. You should get something like this:



```python
import pandas as pd
from scipy import stats
import numpy as np

fraction=np.arange(0.00,0.1,0.01)
z_scores=np.arange(0, 3.1, 0.1)
zDist=dict()
for f in fraction:
    row=[]
    for z in z_scores:
        row.append(1-round(stats.norm.cdf(z+f),4))
    zDist.update({f:row})
pd.DataFrame(zDist)
```

#### EXERCISE 2. 
The cycle time for trucks hauling concrete to a highway construction site is uniformly distributed over the interval 50 to 70 minutes. What is the probability that the cycle time exceeds 65 minutes  if it is known that the cycle time exceeds 55 minutes?


```python
(a, b) = (50, 70)
unifDist = stats.uniform(loc=a, scale= b-a)
# Event A: exceed 65
pA=1-unifDist.cdf(65)

# Event B: exceed 55
pB=1-unifDist.cdf(55)

# p(A | B ) = p(A n B)/p( B )
# p(A n B) = p( A ) 
pAB=pA/pB
pAB
```

#### EXERCISE 3. 
The width of bolts of fabric is normally distributed with mean 950 mm (millimeters) and standard deviation 10 mm.
1.	What is the probability that a randomly chosen bolt has a width of between 947 and 958mm?


```python
mu=950
sigma=10
normDist = stats.norm(mu, sigma)
normDist.cdf(958)-normDist.cdf(947)
```

2.	What is the appropriate value for C such that a randomly chosen bolt has a width less than C with probability .8531?


```python
# ppf(percent point function) aka inverse of ctf(cumulative distribution function) aka Quantile function
normDist.ppf(0.8531)
```

#### EXERCISE 4. 
The school board administered an IQ test to 20 randomly selected teachers. They found that the average IQ score was 114 with a standard deviation of 10. Assume that the cumulative probability is 0.90. What population mean would have produced this sample result?


```python
n=20 # As sample amount < 30, we use t-distribution
x=114 #sample mean
sigma=10
cdf=0.9
```

```python
t=stats.norm.ppf(cdf)
mu=x-t*sigma/(n**0.5)
mu
```
