# Olasılığa Giriş ve Olasılık Dağılımları
Olasılık dağılımları bize, bazı olasılıkların ne şekilde, hangi olasılıklarla gerçekleşebileceği fikrini verir. Örneğin; satışlarımızın belli bir değerden fazla olması olasılığı, 5 günde bir olayın 3 defa gerçekleşmesi olasılığı gibi.
### rassal değişken
Değerlerini bir deneyin sonucundan alan değişkenlere rassal değişken denir. yazı tura atma deneyinin sonuçları.
### dağılım
Evrende meydana gelen olayların ortaya çıkardığı sayısal yapıya dağılım denir.
### olasılık dağılımı
Bir rassal olaya ait değerler ve bu değerlerin gerçekleşme olasılıklarının bir arada ifade edilmesine olasılık dağılmı denir.
### olasılık fonksiyonu
Bir değişkenin herhangi bir değeri alması olasılığını hesaplamaya yarayan fonksiyondur.

## Kesikli ve Sürekli olasılık dağılımları
Elimizdeki veriye göre olasılık dağılımları değişecektir.
### Kesikli Olasılık Dağılımları
- Bernoulli
- Binom
- Poison
### Sürekli Olasılık Dağılımları
- Normal Dağılım
- Uniform Dağılım
- Üstel Dağılım

# Normal Dağılım

## Büyük Sayılar Yasası
Bir rassal değişkenin uzun vadeli kararlılığını tanımlayan olasılık teoremidir.

```python
num = np.random.randint(0, 2, size = 10)
num.mean()
```

```python
#rng = np.random.RandomState(123)
for i in np.arange(1,21):
    deney_sayisi = 2**i
    yazi_turalar = np.random.randint(0, 2, size = deney_sayisi)
    yazi_olasiliklari = np.mean(yazi_turalar)
    print("Atış Sayısı:",deney_sayisi,"---",'Yazı Olasılığı: %.2f' % (yazi_olasiliklari * 100))
```

## Normal Dağılım
- En önemli dağılımdır.
- Normal dağıldığı bilinen sürekli rassal değişkenler için olasılık hesaplaması için kullanılır.
- Belirli sürekli değişkenlerin belirli değerleri alması olasılıkları ile ilgilendiğimiz durumlar için kullanılması uygundur. Örneğin; doların 7 tl üzerine çıkması olasılığı, bir ekonomik göstergenin belli değer aralığında çıkması olasılığı gibi.
- Sürekli rassal olasılıkları integral ile hesaplanır.
- $f(x|μ,σ^2)=\frac{1}{\sqrt{2πσ^2}}e^{-\frac{(x-μ)^2}{2σ^2}}$


# Not: Bu foksiyonun integrelani hesaplayan çılgın matematikçiyi hatırladım; Carl Friedrich Gauss 

## Normal Dağılım Uygulaması: Ürün Satış Olasılıklarının Hesaplanması
- Bir yatırım toplantısı öncesinde gelecek ay ile ilgili satışların belirli değerlerde gerçekleşme olasılıkları hesaplanmak isteniyor.
- Bize bu şirketle ilgili satış olasılıkları lazım. Bu satış olasılıkları üzerinde belirli kararlar alacağız.
- Detaylar:
  - Dağılımın normal olduğu bliniyor. Dağılımın normalliğine görselleştirme yoluyla veya testler( Küçük örneklem için(n<30) Shapiro-Wilk, büyük örneklem için(n>30) Kormoglov-Samirnov testi) yardımıyla karar verilir.
  - Aylık ortalama satış sayısı 80K, standat sapması 5K
- İstenen olasılık: 90K dan fazla satış yapılması olasılığı nedir?

```python
from scipy.stats import norm
```

### norm.cfd(hesaplanmak istenen değer, ortalama, standart sapma)

cdf kümülatif yoğunluk fonksiyonu: Kesikli verilerden farklı olarak burda bir aralığın olasılığı hesaplanacaktır. Bu sebeple kümülatif yoğunluk olarak hesaplanıyor.
cdf(90,80,5): 90 değerine kadar eğri altındaki alanın olasılığı, 80 ortalama ile ve 5 std olarak hesaplanacak
1-norm.cdf(90,80,5): 1 den çıkartılmasının sebebi;norm.cdf(9,80,5) 90 değerine kadar olan olasılığı hesaplar. Ama bizim hesaplamak istediğimiz 90 dan büyük değerler için olasılık hesabıdır. Tüm eğri altındaki alan 1 olduğundan dolayı 1 den çıkartıyoruz.

```python
# 90 dan fazla olması olasılığı
1-norm.cdf(90,80,5)
```

```python
# 73 den az olması olasılığı
norm.cdf(73,80,5)
```

```python
#70 den fazla olması olasılığı
1-norm.cdf(70,80,5)
```

```python
# 85 ile 90 arasında gelmesi olasılığı
norm.cdf(90,80,5)-norm.cdf(85,80,5)
```
