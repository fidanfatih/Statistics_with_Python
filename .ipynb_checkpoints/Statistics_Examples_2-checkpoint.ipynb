```python
import random
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
random.seed(101)
```

#### Example-1

```python
# H0: mu=10 # null hypothesis, (we can say H0: mu  10 )
# H1: mu>10 # alternative hypothesis, right tail
mu0=10
sigma=1.5
n=40
x=10.5
alpha=0.05 #significant level
```

```python
# n >30, sigma is known, observations are independent so we use z test.
z= (x-mu0)/(sigma/sqrt(n))
z
```

```python
pV=1-stats.norm.cdf(z)
pV
```

```python
# P( z>2.1 | H0=True ) =0.0179

# reject the null hypothesis(H0) when p-Value  alpha
# pV < alpha, so We reject the null hypothesis
```

```python
# Critical Value=?
# z= (x-mu)/(sigma/sqrt(n)) >> Xc = mu0 + Z*(sigma/sqrt(n))

mu0 + stats.norm.ppf(1-alpha)*(sigma/sqrt(n))
```

#### Example-3

```python
# H0: mu  110 # null hypothesis
# H1: mu < 110 # alternative hypothesis, (left tail)
mu0=110
s=10
n=20
x=108
alpha=0.01 #significant level
```

```python
# n <30, sigma is unknown, observations are independent so we use t test.
# s=np.std(data)*sqrt(n/df) #Standard Deviation of Sample
# sm=s/sqrt(n) # Standard Error of Sample
# t=(x-mu) / sm
sm=s/sqrt(n)
t=t=(x-mu0) / sm
t
```

```python
df=n-1
tc = stats.t.ppf(alpha,df) # Critical t Value
tc
```

```python
# Since t > tc, We fail to reject the null hypothesis
```

```python
pV=stats.t.cdf(t,df)
pV
```

```python
# P( t< -0.894 | H0=True ) =0.191

# reject the null hypothesis(H0) when p-Value  alpha
# pV > alpha, so We fail to reject the null hypothesis

#conclusion: The principle should accept her original hypothesis. 
# We dont have enough evidence to accept the alternative hypothesis.
```

```python
# Critical Value=?
# t= (x-mu)/ sm >> Xc = mu0 + T*sm

mu0 + stats.t.ppf(alpha,df)*sm
```
